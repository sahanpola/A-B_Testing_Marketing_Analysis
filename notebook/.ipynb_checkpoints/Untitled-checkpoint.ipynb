```python
# import pandas library
import pandas as pd
```

```python
# Upload csv data fFile 
df = pd.read_csv("E:/Data Science/Data Science Project/AB Testing/1.Marketing/Data set/marketing_AB.csv")
```

```python
# Basic Information
print(df.head())
print(df.info)
print(df.describe(include="all"))
```

```python
#Missing Values 
print(df.isnull().sum())
```

```python
# Group Size 
print(df["test group"].value_counts())
```

```python
# Conversion counts
print(pd.crosstab(df["test group"], df["converted"]))
```

```python
print(df.groupby("test group")["converted"].mean())
```

```python
# Check if user IDs are unique
df["user id"].nunique()

# Check duplicate rows
df.duplicated().sum()

# Check duplicate user IDs
df["user id"].duplicated().sum()
```

```python
import numpy as np

# observed difference
ad_conversion = df[df["test group"]=="ad"]["converted"].mean()
psa_conversion = df[df["test group"]=="psa"]["converted"].mean()

observed_difference = ad_conversion - psa_conversion

observed_difference
```

```python
n_permutations = 10000

differences = []

conversion = df["converted"].values
groups = df["test group"].values


for i in range(n_permutations):

    # shuffle group labels
    shuffled_groups = np.random.permutation(groups)

    shuffled_ad = conversion[shuffled_groups=="ad"]
    shuffled_psa = conversion[shuffled_groups=="psa"]

    diff = shuffled_ad.mean() - shuffled_psa.mean()

    differences.append(diff)


differences = np.array(differences)
```

```python
p_value = np.mean(
    differences >= observed_difference
)

print(p_value)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.hist(differences, bins=50)

plt.xlabel("Difference in Conversion Rate")
plt.ylabel("Frequency")
plt.title("Permutation Null Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(differences, bins=50)

plt.axvline(
    observed_difference,
    color="red",
    linewidth=3,
    label=f"Observed Difference = {observed_difference:.5f}"
)

plt.legend()

plt.xlabel("Difference")
plt.ylabel("Frequency")
plt.title("Permutation Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

critical_value = np.percentile(differences, 95)
p_value = np.mean(differences >= observed_difference)

plt.figure(figsize=(12,6))

plt.hist(
    differences,
    bins=50,
    edgecolor="black"
)

plt.axvline(
    observed_difference,
    linewidth=3,
    label=f"Observed Difference = {observed_difference:.5f}"
)

plt.axvline(
    critical_value,
    linestyle="--",
    linewidth=3,
    label=f"95th Percentile = {critical_value:.5f}"
)

plt.axvspan(
    observed_difference,
    differences.max(),
    alpha=0.25,
    label="p-value Region"
)

plt.xlabel("Difference in Conversion Rate")
plt.ylabel("Frequency")
plt.title(f"Permutation Test\np-value = {p_value:.6f}")

plt.legend()
plt.show()
```
